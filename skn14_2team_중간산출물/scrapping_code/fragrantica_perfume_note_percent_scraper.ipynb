```python
# 셀 1: 필요한 모듈 불러오기
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from urllib.parse import quote_plus
from bs4 import BeautifulSoup
import time

# 셀 3: 드라이버 생성 함수
def build_driver(headless=False):
    chrome_options = Options()
    if headless:
        chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    chrome_options.add_argument('--disable-blink-features=AutomationControlled')
    return webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# 셀 4: 향수 하나당 새 크롬 창 열기 + URL 파싱
def get_url_single_run(perfume_name):
    driver = build_driver(headless=False)  # 반드시 False
    search_url = f"https://www.fragrantica.com/search/?query={quote_plus(perfume_name)}"
    driver.get(search_url)
    time.sleep(3)  # 로딩 충분히 기다리기

    try:
        soup = BeautifulSoup(driver.page_source, "html.parser")
        a_tag = soup.select_one(".cell.card.fr-news-box a[href*='/perfume/']")
        href = a_tag["href"] if a_tag else None
        driver.quit()
        return href
    except Exception:
        driver.quit()
        return None

# 셀 5: 전체 향수 순회
detail_urls = []
for name in perfume_names_english:
    url = get_url_single_run(name)
    detail_urls.append((name, url))
    print(f"{name} → {url}")
    time.sleep(1.5)  # 서버에 부담 주지 않도록

# 셀 6: 저장
with open("fragrantica_detail_urls_list.py", "w", encoding="utf-8") as f:
    f.write("detail_urls = [\n")
    for name, url in detail_urls:
        f.write(f"    ({name!r}, {url!r}),\n")
    f.write("]\n")

print("\n✅ 저장 완료 → fragrantica_detail_urls_list.py")
```

```python
# 셀 1) Imports 및 초기 함수들
import re, time, csv
from tqdm.notebook import tqdm

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

from pathlib import Path

# fragrantica_detail_urls_list.py 불러오기
import importlib, fragrantica_detail_urls_list as fdl
importlib.reload(fdl)       # 내용 갱신
detail_urls = fdl.detail_urls

def parse_width_percent(style_text: str):
    m = re.search(r"width:\s*([\d.]+)%", style_text)
    return float(m.group(1)) if m else 0.0


def build_driver(headless: bool = False) -> webdriver.Chrome:
    opts = Options()
    if headless:
        opts.add_argument("--headless=new")
    opts.add_argument("--no-sandbox")
    opts.add_argument("--disable-dev-shm-usage")
    opts.add_argument("--disable-gpu")
    opts.add_argument("--window-size=1400,2200")
    opts.add_experimental_option("excludeSwitches", ["enable-automation"])
    opts.add_experimental_option("useAutomationExtension", False)
    opts.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/124.0.0.0 Safari/537.36")
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=opts)
    driver.set_page_load_timeout(45)
    driver.implicitly_wait(2)
    return driver


def maybe_close_consents(driver, timeout=4):
    selectors = [
        "button[aria-label*='accept' i]",
        ".cc-accept, .accept, button.cookie-accept",
        "#onetrust-accept-btn-handler",
        ".ot-sdk-container button[aria-label*='accept' i]",
    ]
    end = time.time() + timeout
    while time.time() < end:
        for sel in selectors:
            els = driver.find_elements(By.CSS_SELECTOR, sel)
            if els:
                try:
                    driver.execute_script("arguments[0].click();", els[0])
                    return
                except: pass
        time.sleep(0.2)


def wait_accords_ready(driver, wait_sec=15) -> bool:
    try:
        WebDriverWait(driver, wait_sec).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, ".accord-bar"))
        )
        return True
    except TimeoutException:
        return False


def extract_main_accords_selenium(driver):
    bars = driver.find_elements(By.CSS_SELECTOR, ".accord-bar")
    out = []
    for b in bars:
        note = (b.text or "").strip()
        style = b.get_attribute("style") or ""
        w = parse_width_percent(style)
        if note and w:
            out.append((note, w))
    return out


def extract_season_and_time_scores(driver) -> tuple[str, str]:
    try:
        legends = driver.find_elements(By.CSS_SELECTOR, ".vote-button-legend")
        bars = driver.find_elements(By.CSS_SELECTOR, ".voting-small-chart-size > div > div")

        labels = [l.text.strip().lower() for l in legends]
        percents = [parse_width_percent(b.get_attribute("style")) for b in bars]

        paired = list(zip(labels, percents))
        seasons = [f"{k}({v:.1f})" for k, v in paired if k in ["winter", "spring", "summer", "fall"]]
        times = [f"{k}({v:.1f})" for k, v in paired if k in ["day", "night"]]

        return " / ".join(seasons), " / ".join(times)
    except Exception:
        return "", ""


# 셀 2) Scraper Core

def write_csv_header(path: str):
    with open(path, "w", newline="", encoding="utf-8") as f:
        w = csv.DictWriter(f, fieldnames=["perfume_name", "notes_score", "season_score", "day_night_score"])
        w.writeheader()


def append_row(path: str, row: dict):
    with open(path, "a", newline="", encoding="utf-8") as f:
        w = csv.DictWriter(f, fieldnames=["perfume_name", "notes_score", "season_score", "day_night_score"])
        w.writerow(row)


def scrape_detail_urls(detail_urls, out_csv="fragrantica_perfume_summaries.csv",
                       headless=False, pause=0.8, overwrite=True):
    if overwrite or not Path(out_csv).exists():
        write_csv_header(out_csv)

    total_rows = 0

    for name, url in tqdm(detail_urls, desc="Detail pages", unit="page"):
        try:
            driver = build_driver(headless=headless)
            driver.get(url)
            maybe_close_consents(driver)

            ready = wait_accords_ready(driver, wait_sec=20)
            if not ready:
                print(f"[WARN] main accords 미등장: {name} -> {url}")

            accords = extract_main_accords_selenium(driver)
            season_score, day_night_score = extract_season_and_time_scores(driver)

            notes_str = " / ".join([f"{n}({w:.1f})" for n, w in accords])
            row = {
                "perfume_name": name,
                "notes_score": notes_str,
                "season_score": season_score,
                "day_night_score": day_night_score
            }

            append_row(out_csv, row)
            total_rows += 1
            driver.quit()
            time.sleep(pause)

        except Exception as e:
            print(f"[WARN] 실패: {name} -> {url} | {e}")

    print(f"\n✅ Saved → {out_csv} | total perfumes: {total_rows}")
```
